```python
import numpy as np
import pandas as pd
import pprint
```

```python
df = pd.read_csv('data/Export03Mar24.csv', encoding='latin-1', delimiter=';')
```

```python
df.head(10)
```

```python
def get_lift_event_history(current_time, time_window=15):
    current_time = pd.to_datetime(current_time)
    end_time = current_time - np.timedelta64(time_window, 'm')

    lift_util = {}
    for lift in df.Lift.unique():
        lift_df = df.loc[df.Lift == lift]
        lift_df.SteckDatumZeit = pd.to_datetime(lift_df.SteckDatumZeit)
        count = lift_df.SteckDatumZeit.between(end_time, current_time).sum()
        lift_util[lift] = count

    return lift_util
```

```python
my_dict = {}
time = np.datetime64('2024-03-03 07:00:00')
end_time = np.datetime64('2024-03-03 17:00:00')
while True:
    time = time + np.timedelta64(15, 'm')
    my_dict[time] = get_lift_event_history(time)

    if time == end_time:
        break

```

```python
def get_foerder_kappa_all():
    fkappa_df = pd.read_excel("Kurzbeschreibung und Anlagenzuordnung.xlsx")
    foerder_dict = {}

    for i, row in fkappa_df.iterrows():
        foerder_dict[row.Anlage] = {"theoretical_foerder_kapa": row.theor_foerder_kapa_h,
                                    "practical_foerder_kapa": row.practical_foerder_kapa_h}

    return foerder_dict
```

```python
get_foerder_kappa_all()
```

```python
def get_foerder_kappa_specific(lift_name: str):
    fkappa_df = pd.read_excel("data/Kurzbeschreibung und Anlagenzuordnung.xlsx")
    item = fkappa_df.loc[fkappa_df.Anlage == lift_name]
    return {item['Anlage'].values[0]: {"theoretical_foerder_kapa": item.theor_foerder_kapa_h.values[0],
                          "practical_foerder_kapa": item.practical_foerder_kapa_h.values[0]}}
```

# Arriving Departure

```python
import csv 

def read_csv_return_json(csvFilePath):
    
    # create a dictionary
    data = []
    
    # Open a csv reader called DictReader
    with open(csvFilePath, encoding='utf-8-sig') as csvf:
        csvReader = csv.DictReader(csvf)
        
        # Convert each row into a dictionary 
        # and add it to data
        for rows in csvReader:
            data.append(rows)

    # Open a json writer, and use the json.dumps() 
    # function to dump data
    return data
```

```python
read_csv_return_json("data/2024-03-03-Serffaus/Serfaus-Fiss-Ladis - 2024-03-03 (6 - 21h) - Arriving and Departing Visitors.csv")
```
